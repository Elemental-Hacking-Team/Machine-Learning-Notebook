## Primera red neuronal simple con Keras

En este tutorial seguiremos utilizando el conjunto de datos de Pima Indians Diabetes, el cual usaremos para entrenar una red neuronal. Esta técnica es una de las más utilizadas para resolver problemas de machine learning. En [esta nota](https://www.datacamp.com/community/tutorials/deep-learning-python) de DataCamp se encuentra una extensa explicación sobre las redes neuronales y la librería [Keras](https://keras.io).

Primero tenemos que importar los modelos Sequential, y las capas tipo Dense. Con la librería Numpy se importan los datos directamente como array usando el comando np.loadtxt, seleccionando la "," como separador. En este punto ya conocemos bien la estructura de este conjunto de datos utilizados en los tutoriales 2.0 y 3.0. 

En algunas secciones se imprimen mensajes de advertencia, solo hay que ignorarlas ya que no afectan el desarrollo del código.

```python
from keras.models import Sequential
from keras.layers import Dense
import numpy as np

dataset = np.loadtxt("pima-indians-diabetes.csv", delimiter=",")
print(dataset)
```

Como es usual, separamos las salidas de las entradas, agarramos todas las filas del array dataset, con las columnas de 0 a -1 (son contar el -1 que es la última columna) para las X, y tomando la columna final para las y.

```python
X = dataset[:,0:-1]
y = dataset[:,-1]
```

Ahora armamos un modelo tipo Sequential. Un modelo secuencial es apropiado para un conjunto simple de capas donde cada capa tiene exactamente un tensor de entrada y un tensor de salida.

Empezamos agregando una capa de entrada con un número de neuronas igual al número de entradas del conjunto de datos y que tendrá una funcón de activación rectificada lineal, para aprender sobre las funciones de activación pueden usar [este enlace](https://towardsdatascience.com/activation-functions-neural-networks-1cbd9f8d91d6). 

Agregaremos una capa oculta con 12 neuronas, este es un número arbitrario para hacer una primera prueba, una de las maneras de optimizar un modelo podemos probar con diferentes valores para las neuronas de esta capa oculta, lo que se conoce como [análisis de hiperparámetros](https://towardsdatascience.com/hyperparameter-tuning-c5619e7e6624). En otro tutorial veremos como hacer esto de manera sencilla.

Al final agregamos una capa de salida con una neurona, con la función de activación sigmoidea, que regresa valores entre 0 y 1, o en otras palabras, una probabilidad de que la salida sea 1.

```python
model = Sequential()
model.add(Dense(8, input_dim=X.shape[1], activation='relu'))
model.add(Dense(12, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
print(model.summary())
```

Ahora se compila un modelo con la forma que definimos arriba y asignando valores a algunos parámetros para el modelo. La [pérdida](https://keras.io/api/losses/) con binary crossentropy que es adecuada para predicciones entre 0 y 1 como este problema. Un [optimizador](https://keras.io/api/optimizers/) tradicional y la [métrica](https://keras.io/api/metrics/) de exactitud para revisar el desempeño del modelo. 

Es importante destacar que todos estos parámetros pueden generar diferentes resultados al usar diferentes valores y la mejor opción para cada uno depende totalmente del problema que se está analizando. En el texto se encuentra enlaces a notas con información al respecto y los diferentes valores que se pueden utilizar.

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

Paso seguido entrenamos el modelo, aquí se llevará una buena cantidad de tiempo, que depende totalmente de las especificaciones técnicas de nuestro equipo de cómputo. Definiremos un número cualquiera para las epochs y batch, en [esta nota](https://machinelearningmastery.com/difference-between-a-batch-and-an-epoch/) hay información sobre que representa cada uno de esos elementos.

```python
model.fit(X, y, epochs=250, batch_size=10)
```

Ahora evaluamos el modelo, en este caso hemos obtenido una exactitud de 80.08%, un poco más que en el tutorial anterior. Con este resultado podemos esperar una mejora si se hace un ajuste de parámetros ya que las redes tuvieron mejor desempeño que los métodos anteriores.

Nota: Debido a que es un proceso aletario los resultados del modelo y su evaluación pueden variar.

```python
scores = model.evaluate(X, y)
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```

Hacemos un cálculo de las predicciones y después las redondeamos porque se han generado valores entre 0 y 1, cuando nuestra saldida es de 0 ó 1. En este caso se quiere hacer una comparación directa y por eso se redondean los resultados.

```python
predictions = model.predict(X)
print(predictions)
```

```python
rounded = [round(x[0]) for x in predictions]
print(rounded)
```

También se puede usar predict_classes para obtener el resultados directo de cero o uno.

```python
predictions = model.predict_classes(X)
print(predictions)
```

Aquí imprimos los primero 10 casos y vemos la salida obtenida por el modelo y la que se esperaba obtener (valor real)

```python
for i in range(10):
	print('%s => %d (expected %d)' % (X[i].tolist(), predictions[i], y[i]))
```
